```
print("*****************************************\n ------- ARMSTRONG NUMBER CHECKER -------\n*****************************************")
check=True
while check: 
  int_num=[]
  punctuation=["/","*","-","+",".",","]
  sum=0

  num=input("Enter a positive integer number (Press q for exit):")
  if num=="q": 
    print("Thank you so much. Come again !")
    break
  for i in num:
    if (i not in punctuation) and (i.isnumeric()):
      int_num.append(i)
    else:    
      print("It is an invalid entry. Don't use non-numeric, float, or negative values! Try again please...")
      check=False
      break
  if check:
    for i in int_num:
      sum += int(i)**len(int_num)
    if sum==int("".join(int_num)):
      print(sum, "is an Armstrong Number !!!")
    else: 
      print(int("".join(int_num)), "is not an Armstrong Number.")
  else: check=True

```

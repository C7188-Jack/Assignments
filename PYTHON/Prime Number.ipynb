```
print("*****************************************\n ------- PRIME NUMBER CHECKER -------\n*****************************************")
check=True
while check: 
  int_num=[]
  punctuation=["/","*","-","+",".",","]

  num=input("Enter a positive integer number (Press q for exit):")
  if num.lower()=="q": 
    print("Thank you so much. Come again !")
    break
  for i in num:
    if (i not in punctuation) and (i.isnumeric()):
      int_num.append(i)
    else:    
      print("It is an invalid entry. Don't use non-numeric, float, or negative values! Try again please...")
      check=False
      break
  if check:
    num=int("".join(int_num))
    for i in range(2,num):
      if num % i ==0:
        print(num, "is not a prime number :(")
        break
      else:
        if i==num-1:
          print(num,"is a prime number :)")
    
  else: check=True

```

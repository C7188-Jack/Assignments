```
age = input("Are you a cigarette addict older than 75 years old? Yes/No : ").lower()
chronic = input("Do you have a severe chronic disease? Yes/No : ").lower()
immune = input("Is your immune system too weak?").lower()

if age=="yes": age=True
elif age=="no": age=False
else :print("Invalid age data.Please try again.")

if chronic=="yes": age=True
elif chronic=="no": age=False
else :print("Invalid chronic data.Please try again.")


if immune=="yes": age=True
elif immune=="no": age=False
else :print("Invalid immune data.Please try again.")

risk= (age==True or immune==True or chronic==True)
if (risk): print("You are in risky group!")

else: print("You are not in risky group!")
```
